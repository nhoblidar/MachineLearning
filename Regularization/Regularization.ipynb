#Regularization



## Why We Need Regularization

---

##  Problem
- Complex models (many parameters) can **overfit** the data.  
- They memorize noise â†’ poor performance on new data.  

---

##  Solution: Regularization
- Adds a **penalty to large weights** in the cost function.  
- Intuition:  
  - Large weights â†’ model fits noise aggressively  
  - Regularization keeps weights smaller â†’ simpler model  

---

##  Types of Regularization

##  Underfitting
- **Definition:** Model is too simple to capture the patterns in the data.  
- **Causes:**  
  - Too few features  
  - Model is too simple (e.g., linear model for non-linear data)  
  - Insufficient training  
- **Symptoms:**  
  - High training error  
  - High test error  
- **Example:**  
  - Fitting a straight line to curved data  




- **High Bias:** The model is too simple to capture the underlying patterns in the data. For example, using linear regression assumes that features have a linear relationship with the target. If the true relationship is non-linear, the model will consistently make large errors, resulting in underfitting.  

- **Low Variance:** The model is stable with respect to changes in the training data. Small variations in the dataset do not significantly affect the modelâ€™s predictions or the best-fit line, meaning the modelâ€™s outputs are consistent across different training sets.

##  Overfitting
- **Definition:** Model is too complex and learns not just the true patterns but also the **noise** in the training data.  
- **Causes:**  
  - Too many features  
  - Model is too flexible (e.g., high-degree polynomial)  
  - Too little data  
- **Symptoms:**  
  - Very low training error  
  - Very high test error (poor performance on unseen data)  
- **Example:**  
  - A 10th-degree polynomial passing through every training point


- **Low Bias:** The model is flexible enough to capture the underlying patterns in the data. It makes predictions that are generally accurate on the training set.  

- **High Variance:** The model is highly sensitive to small changes in the training data. Even minor variations or outliers can significantly alter the modelâ€™s predictions or the line of best fit, which can lead to overfitting.

## Generalization
- **Definition:** Ability of a model to perform well on **unseen data**.  
- **Characteristics of good generalization:**  
  - Low training error  
  - Low test error  
- **Goal:** Build models that generalize well, not just memorize training data.  

- **Low Bias, Low Variance:** The model achieves a balance between flexibility and stability. It captures the underlying patterns in the data without being overly sensitive to noise or outliers.

This represents the â€œsweet spot,â€ where the model neither underfits nor overfits, resulting in good generalization on unseen data.

## Summary
- **Underfit:** model too simple â†’ high error everywhere  
- **Overfit:** model too complex â†’ memorizes training set but fails test set  
- **Generalization:** sweet spot â†’ model works well on unseen data  
- **Regularization:** helps achieve generalization by penalizing overly complex models
